### Importing Libraries

Before starting, ensure that the pandas and numpy libraries are imported. Python libraries are a collection of external methods that allow its users to do different tasks that are not possible through built in functions. The `pandas` library provides essential tools for data manipulation and the `numpy` library allows for mathematical operations. To do this, type the following at the beginning of your Python script or Jupyter notebook:



```python
import pandas as pd
import numpy as np
```

### Reading the CSV File

1. Locate your CSV file, typically named `CountyHealthData_2014-2015.csv` or similar. Hover your mouse over the file and click the three dots that appear to copy its file path.
2. Create a variable, such as `read`, to store the data read from the CSV file. Use the `pd.read_csv()` function from the pandas library, inserting the copied file path inside the parentheses. Your code will look like this:

```python
read = pd.read_csv('/content/CountyHealthData_2014-2015.csv')
```

```python
read
```

### Accessing a Subset of the Data

1. To focus on data for a specific state, such as North Carolina (NC), use brackets and boolean selectors. Start by typing `read[read['State'] == 'NC']` to filter out data only for NC.
2. Create a copy of this subset to avoid modifying the original data frame. Assign this copy to a new variable, for instance, `NC`, as shown below:

```python
read[read['State'] == 'NC']
```

> Doing this isolates NC from the rest of the data set, which allows for only working with data related to that specific state


```python
NC = read[read['State'] == 'NC'].copy()
```

> Use `.copy()` to copy the subset and assign it to a new variable (NC)

```python
NC
```

### Extracting Specific Columns

1. Decide on the columns you wish to analyze, such as 'Median household income', 'Health care costs', and 'Could not see doctor due to cost'.
2. Create a new variable, like `NC_Health_Costs_Data`, and assign the selected columns to it. This can be done by typing the column names within square brackets after the `NC` variable. For example:

```python
NC_Health_Costs_Data = NC[["State", "Year", "County", "Median household income", "Health care costs", "Could not see doctor due to cost"]]
```

> It is very important to put in the exact names of the columns within quotes and separated with commas, as it would not work otherwise. These column names need to be enclosed within the brackets relative to the previously created variable (NC). `NC[['','']]`

```python
NC_Health_Costs_Data
```

### Exporting the Data to a New CSV File

1. To save your new subset of data, use the `.to_csv()` function. Choose a file name for your new CSV file, such as 'NC_Health_Costs_Data.csv'.
2. You can choose to exclude the index from the CSV by setting `index=False`. The final line of code will be:

```python
NC_Health_Costs_Data.to_csv("NC_Health_Costs_Data.csv", index=False)
```

### Bonus Example: Plotting a Sample of the Dataset

- To visualize a small part of your data, you can use the `matplotlib` library for plotting. First, import matplotlib's `pyplot` module:

```python
import matplotlib.pyplot as plt
```

- First choose the size of the entire graph you want to make in terms of (rows, columns) big using the `plt.figure(figsize=(,)` method. Then, create a simple plot, like a scatter plot, to visualize this data using `plt.scatter()` with specificed columns that were chosen through brackets in the previous cells. The labels and title could be added through `plt.xlabel()`, `plt.ylabel()` , and `plt.title()` functions from the matplotlib.pyplot library. Lastly, if you want to show the grid, you could use `plt.grid(True)` and to show the actual plotted data, use `plt.show()`.  For example, if you want to plot 'Median household income' against 'Health Care Costs in NC' to see correlations:

```python
plt.figure(figsize=(10, 6))
plt.scatter(NC['Median household income'], NC['Health care costs'], alpha=0.7)
plt.title('Median Household Income vs Health Care Costs in NC')
plt.xlabel('Median Household Income ($)')
plt.ylabel('Health Care Costs')
plt.grid(True)
plt.show()
```

> The alpha in `plt.scatter()` controls the opacity of the dots.
